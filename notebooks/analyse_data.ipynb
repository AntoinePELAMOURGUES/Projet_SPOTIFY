```python
import duckdb
import pandas as pd
duckdb.install_extension('parquet')
import os
```

```python
# Connexion à DuckDB
conn = duckdb.connect()

# Requete et affichage de notre premiere Playlist au foramt df
playlist_df = conn.execute("SELECT UNNEST(playlists , recursive := true ) FROM read_json_auto('/mnt/h/spotify_million_playlist_dataset/data/mpd.slice.0-999.json', maximum_object_size=40000000)").df()

playlist_df.head()
```

```python
# Requete afin de créer une table Palylists
conn.execute("CREATE TABLE playlists AS SELECT UNNEST(playlists , recursive:= true) FROM read_json_auto('/mnt/c/Users/antoi/Downloads/spotify_million_playlist_dataset/data/mpd.slice*.json', maximum_object_size = 40000000)")
```

```python
table_playlist= conn.execute("SELECT * FROM playlists LIMIT(5)").df()

table_playlist.head()
```
